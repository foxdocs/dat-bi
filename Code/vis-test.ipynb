```python
import plotly.express as px
```

## Read Iris Data

```python
# we read an html page and use pandas function, which parses all tables in webpages to a list
tables = pd.read_html('https://raw.githubusercontent.com/foxdocs/dat-bi/refs/heads/main/Data/iris_page.html') 
```

```python
# what is the type of the parsed data?
type(tables)
```

```python
# how many elements does it contain?
len(tables)
```

```python
# list elements are indexed with index starting from 0
df = tables[0]
```

```python
# what is the type of the first element of the list?
type(df)
```

```python
# df = df.drop(columns = ['Dataset Order'])
```

```python
# needs some cleaning
df['Species'] = df['Species'].str.replace('I.Â', '')
```

```python
# to plan the plots, better get to know the ranges of the variables
df.describe()
```

## Try Plotly

```python
fig = px.histogram(df, x="Sepal width", y="Sepal length", color="Species")
fig.show()
```

```python
fig = px.scatter(df, x="Sepal width", y="Sepal length", color="Species")
fig.show()
```

```python
fig = px.scatter(df, x="Sepal width", y="Sepal length", color="Species",
                 marginal_x="box", marginal_y="violin", template="simple_white")
fig.show()
```

```python
fig = px.scatter_matrix(df, dimensions=["Sepal width", "Sepal length", 'Petal length' ,'Petal width'], 
                        color="Species", width=1000, height=500)
fig.show()
```

```python
fig = px.scatter_3d(df, x="Sepal width", y="Sepal length", z='Petal length' , color="Species", size="Sepal length", 
                    hover_name="Species", symbol="Species", 
                    color_discrete_map = {"Joly": "blue", "Bergeron": "green", "Coderre":"red"}, width=1000, height=400)
fig.show()
```

```python
df.describe()
```

```python
fig = px.scatter(df, x="Dataset Order", y="Sepal width", animation_frame="Dataset Order", animation_group="Sepal width", 
           size="Sepal length", color="Species", hover_name="Species", 
           log_x=True, range_x=[1, 150], range_y=[1, 10])
fig.show()
```

```python
fig = px.ecdf(df, x="Sepal width", color="Species")
fig.show()
```
