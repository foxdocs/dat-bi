# DBSCAN Clustering of Dummy Data

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np
```

```python
from sklearn import metrics
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist
from yellowbrick.cluster import SilhouetteVisualizer
```

## Data

```python
# Create a  dataset
X, _ = make_circles(n_samples=500, factor=0.3, noise=.03, random_state=3)
```

```python
X
```

```python
# Plotting
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.title("Data")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

## Model

```python
# Train a DBSCAN model
model = DBSCAN(eps=0.1, min_samples=6)
clusters = model.fit_predict(X)
```

```python
# Plotting
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Concentric Circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
labels = model.labels_
labels
```

## Validation

```python
score = silhouette_score(X, labels, metric='euclidean')
print('Silhouette Score: %.3f' % score)
```

```python
# Visualize the silhouette scores of all points
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
visualizer.fit(X)
visualizer.show()  
```
