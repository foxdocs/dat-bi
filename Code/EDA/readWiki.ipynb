## Read Wikipedia

```python
# function to read any subject in any language and 
# return m sentenses summary, the whole text and n links to related documents
def readWiki(subject, lang, n, m):
    wikipedia.set_lang(lang)
    # initialise the structure for holding the results
    result = []
    meta = {}   

    # find a page
    wpage = wikipedia.page(subject) 

    # get the html code of it - can be used for further parsing
    # phtml = wpage.html()    # returns full html code, may take longer
    phtml = wpage.html
    # list of n linked titles
    plinks = wpage.links[0:n]
    
    psum = wikipedia.summary(subject, sentences=m)
    
    meta = {'url': wpage.url, 'title':wpage.title, 'summary': psum, 'links':plinks, 'html':phtml }

    # get the text of the page
    ptext = wpage.content
    result.append(ptext)

    # get the refered sources
    ref = wpage.references
    
    return meta, result, ref
```

```python
metadata, result, reference = readWiki('Neuro-Symbolic AI', 'en', 3, 2)
```
