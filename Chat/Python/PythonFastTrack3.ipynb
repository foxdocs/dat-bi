## Differences Between Pandas and Numpy

1. In __pandas__ we can work with 1D Series and 2D DataFrame, while __numpy__ can process multi-dimensional ndArrays
2. In DataFrame we use column names (like in sql), in ndArrays slice the data based on indices
3. A DataFrame can hold multiple datatypes in different columns, ndArrays contain elements of the same type

## Series

A `Series` is a one-dimensional labeled array in Pandas, capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.
http://pandas.pydata.org/pandas-docs/stable/dsintro.html#series


```python

# create a Series by passing a list of values, Pandas creates a default integer index
ds = pd.Series([1, 3, 5, np.nan, 'seks', 8.9])
ds
```

```python
# define your own index
ps = pd.Series(['seks','fem','fire'], index=[6,5,4])
ps
```

We will use pandas concat() method [explanation here](https://www.tutorialspoint.com/python_pandas/python_pandas_concatenation.htm)

```python
dp = pd.concat([ds,ps])
dp
```
